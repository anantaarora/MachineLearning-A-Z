```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cross_validation import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
dataset = pd.read_csv("C:/Users/Ananta Arora/Documents/Machine Learning A-Z/Polynomial Regression/Position_Salaries.csv")
```

```python
# Importing the dataset
X = dataset.iloc[:,1:2].values
Y = dataset.iloc[:,2].values
```

```python
#X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0) #testSize is in percentage
# we wiil not train and test because:
# we donot have enough information.
# we want to make accurate prediction .
# we have to take whole dataset 4 training model
```

* Sometimes people bluff about their salary in previous companies so to remove that

* we are taking two models here:
> linear regression model &
> polynomial regression 

why? so that we can compare them 

## Fitting Linear Regression to the dataset

```python
lin_reg = LinearRegression()
lin_reg.fit(X,Y)
```

```python
poly_reg = PolynomialFeatures(degree = 4)
```

```python
X_poly = poly_reg.fit_transform(X)
```

```python
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly,Y)
```

## Visualising tha Linear Regression results:

```python
plt.scatter(X,Y,color = "red")
```

```python
plt.scatter(X,Y,color = "red")
plt.plot(X, lin_reg.predict(X),color = "blue")
plt.title("Truth or Bluff (Linear Regression)")
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

* So we saw that the truth salary is away from blue line 
* we need to make a better model to make the points close to the blue line

```python
plt.scatter(X,Y,color = "red")
plt.plot(X, lin_reg_2.predict(poly_reg.fit_transform(X)),color = "blue")
plt.title("Truth or Bluff (Polynomial Regression)")
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

## we can also check for degree 4 which is accurate :
